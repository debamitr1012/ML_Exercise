# Linear Regression

# Canada per capita income


```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/91983/Downloads/canada.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('percapitaincome')
plt.scatter(df.year,df.percapitaincome,color='blue',marker='*')
```

```python
new_df=df.drop('percapitaincome',axis='columns')
new_df
```

```python
percapitaincome=df.percapitaincome
percapitaincome
```

# Create linear regression object

```python
reg=linear_model.LinearRegression()
reg.fit(new_df,percapitaincome)
```

# Predict canada's per capita income in year 2020

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
